```python
def isSurvived(row):
    sex = row['Sex']
    pclass = row['Pclass']
    # embarked = row['Embarked']
    # return table.loc[sex].loc[pclass].loc[embarked].Survived > 0.5
    return table.loc[sex].loc[pclass].Survived > 0.5
```

```python
df_predict = df_test.copy()
df_predict['Survived'] = df5.apply(isSurvived, axis=1)
```

```python
df_predict = df_predict[['PassengerId', 'Survived']]
df_predict
```
